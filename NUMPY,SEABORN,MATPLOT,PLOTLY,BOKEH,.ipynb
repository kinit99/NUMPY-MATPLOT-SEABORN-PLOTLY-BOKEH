```python
#ANSWER 01.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the dataset
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

df = pd.DataFrame(data)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plotting the points
ax.scatter(df['X'], df['Y'], df['Z'], c='blue', marker='o')

# Add labels and title
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title("3D Scatter Plot of Data Points")

# Show the plot
plt.show()
```

```python
#ANSWER 04.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Data preparation
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(df['X'].values.reshape(x.shape),
                df['Y'].values.reshape(y.shape),
                df['Z'].values.reshape(z.shape),
                cmap='viridis')

# Labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```

```python
#ANSWER 05.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(25)

# Data preparation
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))

# Plot bubbles
plt.scatter(df['GDP'], df['Population'],
            s=df['Population'] * 10,  # Bubble size proportional to population
            alpha=0.5,  # Transparency
            c='blue',   # Color
            edgecolors='w',  # Edge color
            linewidth=2)  # Edge line width

# Add labels
for i, country in enumerate(df['Country']):
    plt.text(df['GDP'][i], df['Population'][i], country,
             fontsize=12, ha='right', va='bottom')

# Labeling the axes
plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Bubble Chart: Population vs GDP')

plt.grid(True)
plt.show()
```

**QUESTION 01.**Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
#ANSWER 01.
import numpy as np
from bokeh.plotting import figure, show, output_notebook

# Prepare data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Display in notebook
output_notebook()

# Create and show plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, line_width=2)
show(p)
```

**QUESTION 02.**Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.




```python
#ANSWER 02.
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_notebook

# Generate random data
np.random.seed(0)
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10
sizes = np.random.rand(100) * 50 + 10  # Bubble sizes
colors = np.random.rand(100)  # Colors

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Display in notebook
output_notebook()

# Create scatter plot
p = figure(title="Scatter Plot with Variable Sizes and Colors",
           x_axis_label='x', y_axis_label='y')

# Add scatter renderer
p.scatter(x='x', y='y', size='sizes', color='colors', alpha=0.6, source=df)

# Show the plot
show(p)
```

```python
#ANSWER 04.
import numpy as np
from bokeh.plotting import figure, show, output_notebook

# Generate data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Display in notebook
output_notebook()

# Create histogram plot
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='blue', line_color='black', alpha=0.7)

# Show plot
show(p)
```

**QUESTION 05.**Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)

x = np.linspace(0, 1, 10)

y = np.linspace(0, 1, 10)

xx, yy = np.meshgrid(x, y)


```python
#ANSWER 05.
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

# Generate data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
xx = xx.flatten()
yy = yy.flatten()
data_heatmap_flat = data_heatmap.flatten()

# Prepare data for Bokeh
from bokeh.models import ColumnDataSource
source = ColumnDataSource(data=dict(x=xx, y=yy, value=data_heatmap_flat))

# Display in notebook
output_notebook()

# Create heatmap plot
p = figure(title="Heatmap", x_axis_label='x', y_axis_label='y',
           x_range=(0, 1), y_range=(0, 1))

# Use a linear color mapper for the heatmap
mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(data_heatmap_flat), high=max(data_heatmap_flat))

# Add rectangles for heatmap
p.rect(x='x', y='y', width=0.1, height=0.1, source=source, line_color=None, fill_color=mapper, alpha=0.7)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show plot
show(p)
```
